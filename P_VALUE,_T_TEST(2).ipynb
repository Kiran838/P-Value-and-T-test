```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import math
np.random.seed(8)
school_ages=stats.poisson.rvs(loc=18,mu=35,size=1500)
classA_ages=stats.poisson.rvs(loc=18,mu=30,size=75)
```

```python
classA_ages.mean()
```

```python
_,p_value=stats.ttest_1samp(a=classA_ages,popmean=school_ages.mean())
```

```python
p_value
```

```python
school_ages.mean()
```

```python
if p_value < 0.05:         #Alpha value = 0.05 OR 5%
  print("Accepting null hypothesis")  #Difference in age
else:
  print("Rejecting null hypothesis")  #No difference in age
```

Hence it is concluded that we are accepting the null hypothesis
